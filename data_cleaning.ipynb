#### Active = Confirmed - Deaths - Recovered

```python
active_calculated_rows = df[df['Active'] != df['Confirmed'] - df['Deaths'] - df['Recovered']]
#active_calculated_rows = active_calculated_rows[~active_calculated_rows['ID'].isin(all_null_rows['ID'])]
active_calculated_rows = active_calculated_rows.dropna(subset=['Active', 'Confirmed', 'Deaths', 'Recovered'])
active_calculated_rows = active_calculated_rows[(active_calculated_rows['Active'] >= 0) & (active_calculated_rows['Confirmed'] >= 0) & (active_calculated_rows['Deaths'] >= 0)
                                                & (active_calculated_rows['Recovered'] >= 0)]
active_calculated_rows
```

```python
active_calculated = active_calculated_rows[['ID']].copy()
active_calculated['Column Name'] = 'Active'
active_calculated['Justification'] = "Active != Confirmed - Deaths - Recovered"
active_calculated.to_csv('polluted.csv', mode='a', header=False, index=False)
```

#### Latitude & Longitude
